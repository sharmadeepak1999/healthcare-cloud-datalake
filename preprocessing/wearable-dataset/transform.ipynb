```python
#copying to make further changes

df2=df.copy()
```

```python
#dropping columns with no significance
df2.drop(columns=['ref', 'study_id'], inplace=True)

#df2.drop(['study_id'], axis=1)
```

```python
#dropping columns with missing values and no significance
df2.drop(columns=['appearance_in_first_grade_kinship'], inplace=True)
```

```python

df2['disease_comment'] = df2['disease_comment'].replace('-', 'Not Applicable')
df2['gender'] = df2['gender'].replace('male', 'MALE')
df2['gender'] = df2['gender'].replace('female', 'FEMALE')
df2['handedness'] = df2['handedness'].replace('right', 'RIGHT')
df2['handedness'] = df2['handedness'].replace('left', 'LEFT')
df2['appearance_in_kinship'] = df2['appearance_in_kinship'].replace('true', 'TRUE')
df2['appearance_in_kinship'] = df2['appearance_in_kinship'].replace('false', 'FALSE')
```
